# Data Cleaning

```python
data.info()
```

```python
data.drop(columns=['Unnamed: 0', 'Url_spotify', 'Uri', 'Url_youtube'], inplace=True, errors='ignore')
print(data.columns)
```

# Top 10 Artists - with the Highest Views on YouTube?

#  What are the most common Album Types on Spotify? How many tracks belong to each album type?

```python

a = data['Album_type'].value_counts()
a
```

Most Common Album Type are:
 
album          
single         
compilation

```python
plt.pie(a, labels= a.index,autopct="%1.1f%%", startangle=60, shadow=True, explode= (0.05,0.05,0.05))
plt.title("Most common Album Types on Spotify")
plt.grid("dark")
plt.show()
```

# How do the Average Views, Likes, and Comments are compared between different Album Types?


```python

d = data.groupby('Album_type',as_index=False)[['Views','Likes','Comments']].mean()
d
```

```python
d.melted = pd.melt(d, id_vars='Album_type',var_name='Attributes',value_name='Total')          ## Unpivot the Data by Melt Func
d.melted
```

```python
plt.figure(figsize=(9,4))

sns.barplot(x= 'Album_type',y= 'Total', hue= 'Attributes', data=d.melted,  )
plt.title("Average Views, Likes, and Comments are compared between different Album Types")
plt.show()
```
